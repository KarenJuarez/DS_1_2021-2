## Iteraciones y condicionales

```python
primo = True
for n in range(1, 10):   # Para n en un vector de 2 a 10 [2,3,4,5,6,7,8,9]
    for x in range(2, n):   # Para x en un vector de 2 a n [2,...,n-2,n-1]
        primo = False   # Condici칩n para cada iteraci칩n
        if n % x == 0:   # Si el m칩dulo de n en x es 0, entonces
            print(n, 'es igual a', x, '*', n/x)   # Imprime esto
            break    # Y deja de iterar
        else:   # Si no se cumple lo de arriba
            primo = True   # Volvemos a definir
    if primo:   # Al final del loop pregunta si es primo
        print(n, 'es un numero primo')   # Imprime el PRIMOOOOOOO!
```

Recuerdan la calculadora de propina? Vamos a generar la funci칩n de la siguiente manera:

```python
def calcula_propina(extra = 0):   # Argumentos
    """Funci칩n que calcula la propina. Pregunta cuanto fue el valor de la cuenta y que tal
    estuvo el consumo. El par치metro extra: int indica el dinero extra que le quisieras dar
    al mesero que se suma en la cuenta final.
    """
    cuenta =  float(input('쮺u치nto es de cuenta?:'))  
    servicio = int(input('쮺칩mo estuvo el servicio? \nElige un n칰mero 2 = bueno, 1 = regular, 0 = malo \n'))
    perc_prop = 'nada'
    if servicio == 0:
        perc_prop = 0.1
    elif servicio == 1:
        perc_prop = 0.15
    elif servicio == 2:
        perc_prop = 0.2
    else:
        print('Revisa que la opci칩n sea la correcta')

    propina =  round(cuenta * perc_prop,3)
    print('Esto es de propina',propina) 
    cuenta =  cuenta + propina + extra
    print('Esto pagaras:',cuenta)
```

## __Orientado a objetos__

```python
class Restaurante():
    
    # primera funcion
    def calcula_propina(extra=0):
        cuenta =  float(input('쮺u치nto es de cuenta?:'))  
        servicio = input('쮺칩mo estuvo el servicio? \nElige un n칰mero 2 = bueno, 1 = regular, 0 = malo \n')
        perc_prop = 'nada'
        if int(servicio) == 0:
            perc_prop = 0.1
        elif int(servicio) == 1:
            perc_prop = 0.15
        elif int(servicio) == 2:
            perc_prop = 0.2
        else:
            print('Revisa que la opci칩n sea la correcta')

        propina =  round(cuenta * perc_prop,3)
        print('Esto es de propina',propina) 
        cuenta =  cuenta + propina + extra
        return print('Esto pagaras:',cuenta)
    
    # segunda funcion
    def potencia(x,n=2):
        return x**n
```

```python
Restaurante.potencia(2,3)
```

```python
Restaurante.calcula_propina()
```

Pueden ser usadas en cualquier lugar donde sea requerido un objeto de tipo funci칩n.

Est치n sint치cticamente restringidas a una sola expresi칩n.

Al igual que las funciones anidadas, las funciones lambda pueden hacer referencia a variables desde el 치mbito que la contiene:
