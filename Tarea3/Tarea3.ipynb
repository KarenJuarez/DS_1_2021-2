8) Junto a Michael Jordan, Larry Bird y Magic Johnson son considerados los mejores jugadores en la historia de la NBA. 쮺u치l de los tres fue nombrado MVP en m치s ocasiones? Haz una gr치fica de barras donde muestres el n칰mero de veces que estos tres monstruos de la duela ganaron el MVP a lo largo de sus carreras. 

**HINT**: para filtrar un *dataframe* que cumpla con varias condicionales se sigue el siguiente formato:

`df[(df.nombre_de_la_columna == condicion1) | (df.nombre_de_la_columna == condicion2)]`

Donde el caracter | es el s칤mbolo que indica *or*.

Otra forma de filtrar un DataFrame que cumpla con varias condicionales es con `.isin([])`. Si nunca haz escuchado sobre 칠l, [aqu칤](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.isin.html) encontrar치s algo que te puede dar una pista de como resolver este ejercicio.

```python
MVP=NBA[(NBA.Player=='Magic Johnson')|(NBA.Player=='Larry Bird')|(NBA.Player=='Michael Jordan')]
MVP.head(10)
```

```python
MVP.Player.value_counts(normalize= False)
```

```python
plt.figure(figsize=(10,5))
MVP.Player.value_counts(normalize= False).plot.bar(rot=0,color='darkcyan')
plt.title("Mejores jugadores en la historia de la NBA", fontsize=15)
plt.ylabel("Nombramientos MVP", fontsize=12)
plt.show()
```

El jugador que m치s veces fue nombrado MVP, fue Michael Jordan
